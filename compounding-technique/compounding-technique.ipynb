```python
# Import the libraries
import matplotlib.pyplot as plt
```

```python
# Create a function to calculate compound growth with contributions
def calculate_compound_growth_with_contributions(principal, target, rate_of_return, annual_contribution):
    balance_over_time = [principal]
    years = 0

    # Create a loop until we hit the target
    while principal < target:
        principal = principal * (1 + rate_of_return) + annual_contribution
        balance_over_time.append(principal)
        years += 1

    return years, balance_over_time
```

```python
# Get the initial investment
initial_investment = 100000

# Get the target amount 
target_amount = 1000000

# Get the annual compunding return rate (as a decimal)
annual_return_rate = 0.07

# Get the annual contributions
annual_contribution = 10000

# Calculate the number of years and balance over time with contributions
years_to_target, balance_over_time = calculate_compound_growth_with_contributions(initial_investment, target_amount, annual_return_rate, annual_contribution)
```

```python
# Plot the inverstment with contributions
plt.figure(figsize=(10,5))
plt.plot(balance_over_time, marker='o', label='Growth with Yearly Contributions')
plt.title("Investment Growth Over Time with Yearly Contributions")
plt.xlabel("Years")
plt.ylabel("Balance (USD)")
plt.grid(True)
plt.legend()
plt.show()

print(f"With an annual contribution of ${annual_contribution}, it will take about {years_to_target} years to grow ${initial_investment} to ${target_amount} with an annual return rate of ${round(annual_return_rate*100, 2)}%.")
```
